```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def step():
    import random
    return 1. if random.random() > .5 else -1
```

```python
def walk(n):
    x = np.zeros(n)
    dx = 1./n
    for i in range(n - 1):
        x_new = x[i] + dx * step()
        if x_new > 5e-3:
            x[i + 1] = 0
        else:
            x[i + 1] = x_new
    return x
n = 100000
x = walk(n)
```

```python
plt.plot(x)
```
